# 02 - Solar Irradiance Analysis

Compute irradiance statistics, seasonal patterns, anomalies, and clearsky indices
using the SolarAnalyzer class.

```python
from solar_intelligence.data_loader import generate_synthetic_solar_data
from solar_intelligence.solar_analysis import SolarAnalyzer
from solar_intelligence.visualization import SolarVisualizer
import holoviews as hv
hv.extension('bokeh')
```

```python
ds = generate_synthetic_solar_data(lat=28.6, lon=77.2, start_year=2020, end_year=2023)
analyzer = SolarAnalyzer(dataset=ds, latitude=28.6, longitude=77.2)
viz = SolarVisualizer(width=700, height=400)
```

## Average Daily Irradiance

```python
avg = analyzer.average_daily_irradiance()
for component, value in avg.items():
    print(f"{component}: {value:.2f} kWh/m2/day")
```

## Monthly Irradiance Patterns

```python
monthly = analyzer.monthly_irradiance()
viz.monthly_irradiance_bar(monthly)
```

## Seasonal Patterns

```python
seasonal = analyzer.seasonal_patterns()
print(seasonal.to_string())
```

## Rolling Average Time Series

```python
rolling = analyzer.rolling_average(window=30)
viz.daily_irradiance_timeseries(rolling)
```

## Anomaly Detection

```python
anomalies = analyzer.anomaly_detection()
print(f"Anomaly range: {anomalies['anomaly'].min():.2f} to {anomalies['anomaly'].max():.2f} kWh/m2/day")
```

## Clearsky Index

```python
csi = analyzer.clearsky_index()
print(csi.to_string(index=False))
```

## Full Summary

```python
summary = analyzer.summary()
for key, value in summary.items():
    if key != 'location':
        print(f"{key}: {value}")
```
