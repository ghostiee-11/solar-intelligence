# 04 - Energy Estimation & Financial Analysis

Estimate PV energy output with temperature derating, compute ROI/NPV/payback,
and visualize financial projections.

```python
from solar_intelligence.data_loader import generate_synthetic_solar_data
from solar_intelligence.energy_estimator import EnergyEstimator
from solar_intelligence.financial import FinancialAnalyzer
from solar_intelligence.visualization import SolarVisualizer
import holoviews as hv
hv.extension('bokeh')
```

```python
ds = generate_synthetic_solar_data(lat=28.6, lon=77.2)
viz = SolarVisualizer(width=700, height=400)
```

## Energy Estimation

```python
estimator = EnergyEstimator(
    panel_efficiency=0.20,
    panel_area=1.7,
    num_panels=20,
    system_losses=0.14,
)
print(f"System capacity: {estimator.system_capacity_kw:.1f} kW")
print(f"Total area: {estimator.total_area:.1f} m2")
```

```python
annual = estimator.estimate_annual_energy(ds)
monthly = estimator.estimate_monthly_energy(ds)
summary = estimator.system_summary(ds)

print(f"Annual energy: {annual:,.0f} kWh")
print(f"Capacity factor: {summary['performance']['capacity_factor_pct']:.1f}%")
print(f"Specific yield: {summary['performance']['specific_yield_kwh_kwp']:.0f} kWh/kWp")
```

## Monthly Energy Projection

```python
viz.energy_projection_area(monthly)
```

## Financial Analysis

```python
fa = FinancialAnalyzer(
    system_cost=20000,
    electricity_rate=0.12,
    incentive_percent=0.30,
    rate_increase=0.03,
)

fin = fa.financial_summary(annual)
print(f"Net cost: ${fin['investment']['net_cost']:,.0f}")
print(f"First year savings: ${fin['returns']['first_year_savings']:,.0f}")
print(f"Payback: {fin['returns']['payback_years']} years")
print(f"NPV (25yr): ${fin['returns']['npv_25yr']:,.0f}")
print(f"ROI: {fin['returns']['roi_pct']}%")
print(f"CO2 offset: {fin['environmental']['annual_co2_offset_kg']:,.0f} kg/year")
print(f"Equivalent trees: {fin['environmental']['equivalent_trees']}")
```

## Payback Timeline

```python
lifetime = fa.lifetime_savings(annual)
viz.payback_timeline(lifetime)
```

## Carbon Savings

```python
viz.carbon_savings_bar(lifetime)
```
